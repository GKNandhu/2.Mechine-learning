```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset=pd.read_csv("50_Startups.csv")
dataset=pd.get_dummies(dataset,drop_first=True)
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend',
       'State_Florida', 'State_New York']]
dependent=dataset[['Profit']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
```

```python
grid_para={'criterion':['mse','mae','friedman_mse'],'splitter':['best','random'],
           'max_features':['auto','sqrt','log2']}
```

```python
grid=GridSearchCV(DecisionTreeRegressor(),grid_para,refit=True,verbose=3,n_jobs=-1)
```

```python
grid.fit(independent,dependent)
```

```python
re=grid.cv_results_
```

```python
print("best r2 value{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
age=float(input("age"))
bmi=float(input("bmi"))
child=float(input("child"))
sex=float(input("male 0 or 1"))
smoker=float(input("smoker yes 0 or 1"))
```

```python
finalvalue=grid.predict([[age,bmi,child,sex,smoker]])
```
